```python
import re
import pandas as pd
from PyPDF2 import PdfReader

rutaArchivo = 'C:/Compilador/Analisis Anual 2018 ETAS.pdf'
documento = PdfReader(rutaArchivo)
dir(documento)

len(documento.pages)
pagina = documento.pages[4]
texto = pagina.extract_text()
texto
```

```python
import re
patron = r'\n\s*(\d+)\s*(.+?)\s{2,}(\w+\s*\w+\s*\w+\s*\w+\s*\w+\s*\w+)\s*(.+?)\s{2,}(\d+)\s*' 
datos = re.findall(patron, texto)


etas = pd.DataFrame(datos, columns=['SE', 'Area de salud', 'ETA Reportada', 'Fuente de contagio', 'Numero de casos'])
etas
```
